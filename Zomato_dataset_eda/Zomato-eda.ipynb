```python
df_country_code=pd.read_excel("Country-Code.xlsx")
```

```python
df.head()
```

```python
final=pd.merge(df,df_country_code,on="Country Code",how='left')
```

```python
final.head()
```

```python
df.dtypes
```

```python
final.Country.value_counts()
```

```python
country_names =final.Country.value_counts().index
```

```python
country_val=final.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:

- India has the largest number of zomato orders. 
- USA and United Kingdom takes the second and third places

```python
final.columns
```

```python
ratings=final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
 ratings
```

## Observation
- Ratings from 4.5 - 4.9 ---> Excellent
- Ratings from 4.0 - 4.5 ---> Very Good
- Ratings from 3.5 - 3.9 ---> Good
- Ratings from 2.5 - 3.4 ---> Average
- Ratings from 1.8 - 2.4 ---> Poor



```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(14,7)

sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```

The maximum number of ratings is between 2.5 to 3.9

```python
sns.countplot(x='Rating color',data=ratings,palette=['grey','red','orange','yellow','green','green'])
```
